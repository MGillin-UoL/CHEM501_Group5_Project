```python
import paho.mqtt.subscribe as subscribe
import pandas as pd
import sqlite3
from datetime import datetime
import time

# List of topics to subscribe to
topics = ['MG_CHEM501_Project_Test1', 'MG_CHEM501_Project_Test2']

# MQTT broker hostname
hostname = "pf-eveoxy0ua6xhtbdyohag.cedalo.cloud"

# Initialize an empty DataFrame for the table
data_table = pd.DataFrame(columns=['Timestamp', 'Topic', 'Value'])

# Database setup
db_filepath = 'mqtt_data.db'  # SQLite database file
table_name = 'mqtt_data'

# Connect to SQLite database (creates the file if it doesn't exist)
conn = sqlite3.connect(db_filepath)
cursor = conn.cursor()

# Create the table if it doesn't exist
cursor.execute(f"""
CREATE TABLE IF NOT EXISTS {table_name} (
    Timestamp TEXT,
    Topic TEXT,
    Value REAL
)
""")
conn.commit()

# Track the last time the database was updated
last_save_time = time.time()

while True:  # Continuously listen for data
    for topic in topics:
        # Subscribe to one topic at a time
        m = subscribe.simple(topic, hostname=hostname, retained=False, msg_count=1)
        
        # Extract data
        try:
            payload = float(m.payload)
            timestamp = datetime.now().isoformat()  # Capture the current time in ISO format
            
            # Append to the DataFrame
            new_row = {'Timestamp': timestamp, 'Topic': topic, 'Value': payload}
            data_table = pd.concat([data_table, pd.DataFrame([new_row])], ignore_index=True)
            
            # Print the most recent entry for monitoring
            print(new_row)
        except ValueError:
            print(f"Error: Could not convert payload from {topic} to float.")
    
    # Save the DataFrame to the SQL database every 30 seconds
    current_time = time.time()
    if current_time - last_save_time >= 30:  # Check if 30 seconds have passed
        # Write the DataFrame to the SQL table
        data_table.to_sql(table_name, conn, if_exists='append', index=False)
        print(f"Data saved to SQL database {db_filepath} at {datetime.now()}")
        
        # Clear the DataFrame after saving to avoid duplicate entries
        data_table = pd.DataFrame(columns=['Timestamp', 'Topic', 'Value'])
        last_save_time = current_time  # Reset the save time
```

```python
import paho.mqtt.subscribe as subscribe
import pandas as pd
import sqlite3
from datetime import datetime
import os

# List of topics to subscribe to
topics = ['MGTemp', 'MGPressure']

# MQTT broker hostname
hostname = "pf-eveoxy0ua6xhtbdyohag.cedalo.cloud"

# Specify the database file path
db_directory = os.path.expanduser("~/mqtt_data")  # Save in the 'mqtt_data' folder in your home directory
os.makedirs(db_directory, exist_ok=True)  # Ensure the directory exists
db_filepath = os.path.join(db_directory, "mqtt_data.db")

# SQLite database connection
conn = sqlite3.connect(db_filepath)
cursor = conn.cursor()

# Create the table if it doesn't exist
cursor.execute("""
CREATE TABLE IF NOT EXISTS mqtt_data (
    Timestamp TEXT,
    Topic TEXT,
    Value REAL
)
""")
conn.commit()

print(f"Database file is located at: {db_filepath}")

while True:  # Continuously listen for data
    for topic in topics:
        try:
            # Subscribe to one topic at a time
            m = subscribe.simple(topic, hostname=hostname, retained=False, msg_count=1)
            payload = float(m.payload)  # Ensure payload is a float
            timestamp = datetime.now().isoformat()  # ISO 8601 format
            
            # Save data to the database immediately
            cursor.execute("""
            INSERT INTO mqtt_data (Timestamp, Topic, Value)
            VALUES (?, ?, ?)
            """, (timestamp, topic, payload))
            conn.commit()
            
            # Print data for monitoring
            print(f"Saved: {timestamp}, {topic}, {payload}")
        
        except ValueError:
            print(f"Error: Could not convert payload from {topic} to float.")
        except Exception as e:
            print(f"Unexpected error: {e}")
```
